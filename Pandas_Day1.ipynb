```python
import pandas as pd
import numpy as np
```

```python
votes = np.array([120, 240, 150])
```

```python
ratings = np.array([4.1, 4.5, 3.9])
restaurants = np.array(['R1', 'R2', 'R3'])
```

```python
#1D Numpy to 1D Pandas

ratings_series= pd.Series(ratings, index=restaurants, name='Ratings') #1 d structure

votes_series= pd.Series(votes, index=restaurants, name='Votes')
```

```python
ratings_series #this is converted from 1D Numpy array
```

```python
votes_series #this is converted from 1D Numpy array
```

```python
#concat column wise
pd.concat([ratings_series,votes_series], axis=1) #axis =1 when you have to attach 2 things column wise
```
